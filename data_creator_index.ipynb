```python
import data_creator
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = 16, 9
plt.rcParams['lines.linewidth'] = 3

font = {'weight' : 'normal',
      'size'   : 17}

plt.rc('font', **font)
#--------------

output_co2_fe = data_creator.data_gen(datafile = 'RCP3PD')
output_co2_fe_RCP85 = data_creator.data_gen(datafile = 'RCP85')
```

```python
y_0 = 1765
time = np.arange(y_0,y_0+736)

### Plotting all the different methods and parameters sets of radiative forcings
#           We plot FAIR standard and MAGICC optimized parameters, using a additive and differencing technique


plotting = 'forcing'

gases = ['CO2','CH4','N2O','other','anthro_0']
version = ['adding_fair','differencing_fair','adding_magicc','differencing_magicc']
colours = ['red','blue','green','orange','black']
linewidths = [3,1,2,6]

for j in range(len(version)):
    for i in range(0,len(gases)):
        if j == 3:
            alpha_val = 0.3
        else:
            alpha_val = 1
        if j == 2:
            linestyles = ':'
        else:
            linestyles = '-'
        plt.plot(time, output_co2_fe[version[j]][plotting][gases[i]], linestyles, color = colours[i], linewidth = linewidths[j], alpha = alpha_val)

plt.grid(True)
plt.xlabel('Year')
plt.ylabel('Radiative forcing (Wm$^{-2}$)')
plt.xlim(1900,2100)
```

```python
### Plotting all the different methods and parameters sets of CO2-fe emissions
#           We plot FAIR standard and MAGICC optimized parameters, using a additive and differencing technique

plotting = 'emissions'
gases = ['CO2','CH4','N2O','other','anthro_0']
version = ['adding_fair','differencing_fair','adding_magicc','differencing_magicc']
colours = ['red','blue','green','orange','black']
linewidths = [3,1,2,6]

for j in range(len(version)):
    for i in range(0,len(gases)):
        if j == 3:
            alpha_val = 0.3
        else:
            alpha_val = 1
        if j == 2:
            linestyles = ':'
        else:
            linestyles = '-'
        plt.plot(time, output_co2_fe[version[j]][plotting][gases[i]], linestyles, color = colours[i], linewidth = linewidths[j], alpha = alpha_val)

plt.grid(True)
plt.xlabel('Year')
plt.ylabel('Annual emissions (GtCO$_2$)')
plt.xlim(1900,2100)
```

```python
### Plotting all the different methods and parameters sets of concentrations
#           We plot FAIR standard and MAGICC optimized parameters, using a additive and differencing technique

plotting = 'concentration'
gases = ['CO2','CH4','N2O','other','anthro_0']
version = ['adding_fair','differencing_fair','adding_magicc','differencing_magicc']
colours = ['red','blue','green','orange','black']
linewidths = [3,1,2,6]

for j in range(len(version)):
    for i in range(0,len(gases)):
        if j == 3:
            alpha_val = 0.3
        else:
            alpha_val = 1
        if j == 2:
            linestyles = ':'
        else:
            linestyles = '-'
        plt.plot(time, output_co2_fe[version[j]][plotting][gases[i]], linestyles, color = colours[i], linewidth = linewidths[j], alpha = alpha_val)

plt.grid(True)
plt.xlabel('Year')
plt.ylabel('CO$_2$-fe concentrations (ppmv)')
plt.xlim(1900,2100)
```

```python
#estimating the cumulative emissions for co2 between 1870 and 2011
#plotting the differences between the RCP3PD and RCP85 annual co2-fe emissions trajectories 1900-2100
#for both FAIR standard and MAGICC parameters

plotting = 'emissions'
gases = ['CO2','CH4','N2O','other','anthro_0']
version = ['adding_fair']
colours = ['red','blue','green','orange','black']
colours2 = ['darkred','darkblue','darkgreen','brown','grey']

fig = plt.figure(figsize= (16,6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

for j in range(len(version)):
    for i in range(0,len(gases)):
        ax1.plot(time, output_co2_fe_RCP85[version[j]][plotting][gases[i]], linestyles, color = colours[i], linewidth = 2, alpha = 0.7)
        ax1.plot(time, output_co2_fe[version[j]][plotting][gases[i]], linestyle = ':', color = colours2[i], linewidth = 3, alpha = 1)

ax1.text(1910,132, 'FAIR, adding')
ax1.grid(True)
ax1.set_xlabel('Year')
ax1.set_ylabel('CO$_2$-fe annual emissions (GtCO$_2$)')
ax1.set_xlim(1900,2100)
ax1.set_ylim(-25,165)

cum_co2_fe_RCP3PD = np.cumsum(output_co2_fe[version[0]]['emissions']['CO2']) * 12./44.
cum_co2_fe_RCP85 = np.cumsum(output_co2_fe_RCP85[version[0]]['emissions']['CO2']) * 12./44.

print 'Cumulative carbon output 1870-2011 inclusive (RCP3PD, FAIR, adding):', cum_co2_fe_RCP3PD[2011-y_0] - cum_co2_fe_RCP3PD[1870-y_0], 'GtC'
print 'Cumulative carbon output 1870-2011 inclusive (RCP85, FAIR, adding):', cum_co2_fe_RCP85[2011-y_0] - cum_co2_fe_RCP85[1870-y_0], 'GtC'

#-------------
version = ['adding_magicc']
for j in range(len(version)):
    for i in range(0,len(gases)):
        ax2.plot(time, output_co2_fe_RCP85[version[j]][plotting][gases[i]], linestyles, color = colours[i], linewidth = 2, alpha = 0.7)
        ax2.plot(time, output_co2_fe[version[j]][plotting][gases[i]], linestyle = ':', color = colours2[i], linewidth = 3, alpha = 1)

ax2.text(1910,132, 'MAGICC, adding')
ax2.grid(True)
ax2.set_xlabel('Year')
ax2.set_ylabel('CO$_2$-fe annual emissions (GtCO$_2$)')
ax2.set_xlim(1900,2100)
ax2.set_ylim(-25,165)

cum_co2_fe_RCP3PD_MAGICC = np.cumsum(output_co2_fe[version[0]]['emissions']['CO2']) * 12./44.
cum_co2_fe_RCP85_MAGICC = np.cumsum(output_co2_fe_RCP85[version[0]]['emissions']['CO2']) * 12./44.

print 'Cumulative carbon output 1870-2011 inclusive (RCP3PD, MAGICC, adding):', cum_co2_fe_RCP3PD_MAGICC[2011-y_0] - cum_co2_fe_RCP3PD_MAGICC[1870-y_0], 'GtC'
print 'Cumulative carbon output 1870-2011 inclusive (RCP85, MAGICC, adding):', cum_co2_fe_RCP85_MAGICC[2011-y_0] - cum_co2_fe_RCP85_MAGICC[1870-y_0], 'GtC'

cum_co2_fe_RCP3PD_FAIR_diff = np.cumsum(output_co2_fe['differencing_fair']['emissions']['CO2']) * 12./44.
cum_co2_fe_RCP85_FAIR_diff = np.cumsum(output_co2_fe_RCP85['differencing_fair']['emissions']['CO2']) * 12./44.
cum_co2_fe_RCP3PD_MAGICC_diff = np.cumsum(output_co2_fe['differencing_magicc']['emissions']['CO2']) * 12./44.
cum_co2_fe_RCP85_MAGICC_diff = np.cumsum(output_co2_fe_RCP85['differencing_magicc']['emissions']['CO2']) * 12./44.

print 'Cumulative carbon output 1870-2011 inclusive (RCP3PD, MAGICC, differencing):', cum_co2_fe_RCP3PD_MAGICC_diff[2011-y_0] - cum_co2_fe_RCP3PD_MAGICC_diff[1870-y_0], 'GtC'
print 'Cumulative carbon output 1870-2011 inclusive (RCP85, MAGICC, differencing):', cum_co2_fe_RCP85_MAGICC_diff[2011-y_0] - cum_co2_fe_RCP85_MAGICC_diff[1870-y_0], 'GtC'
print 'Cumulative carbon output 1870-2011 inclusive (RCP3PD, FAIR, differencing):', cum_co2_fe_RCP3PD_FAIR_diff[2011-y_0] - cum_co2_fe_RCP3PD_FAIR_diff[1870-y_0], 'GtC'
print 'Cumulative carbon output 1870-2011 inclusive (RCP85, FAIR, differencing):', cum_co2_fe_RCP85_FAIR_diff[2011-y_0] - cum_co2_fe_RCP85_FAIR_diff[1870-y_0], 'GtC'
```

```python
fig3 = plt.figure()
ax22 = fig3.add_subplot(111)

ax22.plot(time, output_co2_fe['adding_magicc']['forcing']['anthro_0'], color = 'black', label = 'RF anthro from RCP3PD (MAGICC)', linewidth = 5)
ax22.plot(time, output_co2_fe['adding_fair']['forcing']['anthro_0'], color = 'yellow', label = 'RF anthro from RCP3PD (FAIR)', linewidth = 1.5)
ax22.plot(time, output_co2_fe['adding_magicc']['forcing']['CO2'], color = 'darkred', label = 'CO$_2$ adding (MAGICC)', linewidth = 5)
ax22.plot(time, output_co2_fe['adding_fair']['forcing']['CO2'], color = 'red', label = 'CO$_2$ adding (FAIR)', linewidth = 1.5)
ax22.plot(time, output_co2_fe['adding_magicc']['forcing']['N2O'], color = 'darkgreen', label = 'N$_2$O adding (MAGICC)', linewidth = 5)
ax22.plot(time, output_co2_fe['adding_fair']['forcing']['N2O'], color = 'lime', label = 'N$_2$O adding (FAIR)', linewidth = 1.5)
ax22.plot(time, output_co2_fe['adding_magicc']['forcing']['CH4'], color = 'darkblue', label = 'CH$_4$ adding (MAGICC)', linewidth = 5)
ax22.plot(time, output_co2_fe['adding_fair']['forcing']['CH4'], color = 'cyan', label = 'CH$_4$ adding (FAIR)', linewidth = 1.5)
ax22.plot(time, output_co2_fe['adding_magicc']['forcing']['other'], color = 'orangered', label = 'other adding (MAGICC)', linewidth = 5)
ax22.plot(time, output_co2_fe['adding_fair']['forcing']['other'], color = 'gold', label = 'other adding (FAIR)', linewidth = 1.5)
ax22.grid(True)
ax22.set_xlabel('Year')
ax22.set_ylabel('Radiative forcing (Wm$^{-2}$)')
ax22.legend(loc = 'upper left', framealpha = 1, shadow = False, edgecolor = 'black', fontsize=13)
ax22.set_xlim(1900,2100)
ax22.text(1869,-1.2,'NOTE: No difference is made when plotting anthro_0 with fair or magicc params because it starts at the other_rf stage',fontweight = 'bold')
fig3.suptitle('Radiative forcings forward calculated comparison between different parameter sets', fontweight = 'bold')

#fig3.savefig('./Figures-final/Radiative forcings comparison between different parameter sets.png',dpi=600)
```
